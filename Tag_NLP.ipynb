```python
from pythainlp import word_tokenize, Tokenizer
tk_df = pre_df.copy()
def tokenize(text):
    temp = []
    text = text.replace(' ','')
    custom_tokenizer = Tokenizer(cat_dict)
    text = str(text)
    custom_token = custom_tokenizer.word_tokenize(text)
    custom_token = [x for x in custom_token if x != 'WSNUMBER']
    token = word_tokenize(text, engine="newmm")
    for i in custom_token :
           if i in cat_dict:
                token.append(i)
    return token
tk_df['content'] = tk_df['content'].apply(tokenize)
tk_df['title'] = tk_df['title'].apply(tokenize)
```

```python
from pythainlp.corpus.common import thai_stopwords
st_df = tk_df.copy()
def stop_word(text):
    removedList=[x for x in text if not x in thai_stopwords() ] #remove stop word
    removedList=[x for x in removedList if not x in [' ']] #remove space
    text=' '.join(removedList)
    return text
st_df['content'] = st_df['content'].apply(stop_word)
st_df['title'] = st_df['title'].apply(stop_word)
```
